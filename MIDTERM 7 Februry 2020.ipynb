### ข้อ 1 > สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่ :
  - function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน โดย list แรกเป็น list ของตัวเลข list ที่สองเป็น list ของ text
  - ถ้าใน list แรก เป็นเลขคี่ ให้หาค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
  - ถ้าใน list แรก เป็นเลขคู่ ให้หาค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
  - output ของ function คือ dict ที่สร้างขึ้น
#### เช่น 
 output1 = smartDict([1,2,3],['apple','banana','orange'])
จะได้ output1 เป็น {1:'apple','banana':2,3:'orange'}

 output2 = smartDict([7,1,4,6],['A','B','E','Zzz']) จะได้ output2 เป็น {7:'A',1:'B','E':4,'Zzz':6}

```python
def smartDict(listnumber,listtext) :
    retdict = {}
    for index in range (len(listtext)) :
        if listnumber[index]%2==1 :
                retdict[listnumber[index]] = listtext[index]
        else :
                retdict[listtext[index]] = listnumber[index]
    return retdict
```

```python
output1 = smartDict([1,2,3],['apple','banana','orange'])
output1
```

```python
output2 = smartDict([7,1,4,6],['A','B','E','Zzz'])
output2
```

#### 2.1 กรดอะมิโนใด (A,T,G,C) มีปริมาณมากที่สุดและมีเท่าใด

```python
import pandas as pd
```

```python
DNAseq = pd.read_csv('C:\\DWDM\\Python\\python_dataviz\\Wuhan_nCoV_GNome.txt', header = None)
```

```python
DNAseq
```

```python
def countChar(seq,character) :
      charcount = 0
      for char in seq :
            if char == character :
                charcount = charcount + 1
      return charcount
```

```python
ANum = 0
TNum = 0
GNum = 0
CNum = 0

for row in range(DNAseq.shape[0]):
     ANum += countChar(DNAseq.iloc[row][0],'A')
     TNum += countChar(DNAseq.iloc[row][0],'T')
     GNum += countChar(DNAseq.iloc[row][0],'G')
     CNum += countChar(DNAseq.iloc[row][0],'C')
```

```python
DNA = {}
DNA['A'] = ANum
DNA['T'] = TNum
DNA['G'] = GNum
DNA['C'] = CNum
```

```python
DNA
```

```python
Max = 0
for index in DNA:
    if DNA[index] > Max :
        Max = DNA[index]
for index in DNA:
    if Max == DNA[index] :
        print (f'กรดอะมิโนที่มีปริมาณมากที่สุดคือ {index} และมีปริมาณ {Max}')
    
```

#### 2.2 จงหา sequence ที่มีขนาด 3 และ 4 ที่เกิดขึ้นซ้ำมากที่สุด

```python
def populate3charstring(seq) :
      str_list = list()
      for i in range(len(seq)-4) :
           str_list.append(seq[i:i+3])
      return str_list
```

```python
populate3charstring('ATTAAAGGTTTATACCTTCCCAGGTAACAAACCAACCAACTTTCGATCTCTTGTAGATCTGTTCTCTAAA')
```

```python
list_all_3char = list()
for row in range(DNAseq.shape[0]) :
      list_all_3char = list_all_3char + populate3charstring(DNAseq.iloc[row][0])
len(list_all_3char)
```

```python
unique_3char = list(set(list_all_3char))
len(unique_3char )
```

```python
dict_3char = {}
for sub_str in unique_3char : 
    num = 0 
    for s in list_all_3char : 
        if sub_str == s :
            num += 1
    dict_3char[sub_str] = num
```

```python
dict_3char
```

```python
maxVal3 = -100
maxKey3 = None
for k in  dict_3char.keys() :
    if dict_3char[k] > maxVal3 :
        maxVal3 = dict_3char[k]
        maxKey3 = k
```

```python
maxKey3
```

```python
print(f'sequence ที่มีขนาด 3 คือ {maxKey3} เกิดขึ้นซ้ำกันมากสุดจำนวน {maxVal3}')
```
