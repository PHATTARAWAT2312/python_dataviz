# 1.นำข้อมูลเข้า

```python
data = pd.read_csv('C:\\DWDM\\Python\\python_dataviz\\2019_nCoV_data.csv')
```

# 2. Data cleansing

##### ตรวจสอบค่า missing 

```python
data.isnull().any()
```

```python
data['Province/State']=data['Province/State'].fillna('missing')
data
```

 #### ตรวจสอบค่า missing อีกครั้ง จะเห็นได้ว่าไม่มีค่า missing แล้ว

```python
data.isnull().any() 
```

```python
data_new = data[data['Country']!='missing']
data_new
```

# 3. Groupby ประเทศ

```python
datan = data_new.groupby('Country').max().reset_index()
datan
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
plt.bar(datan.iloc[:,0], datan.iloc[:,5], color = 'b')
plt.title(u'จำนวนผู้ติดเชื้อของแต่ละประเทศ', fontname='Tahoma', fontsize='16', color = 'b')
plt.xlabel('Country', fontname='Tahoma', fontsize='16', color = 'b')
plt.ylabel('Confirmed', fontname='Tahoma', fontsize='16', color = 'b')
```

# 4.จะเห็นได้ว่าในตาราง ค่าของ mainchina มีค่า scale สูงกว่าเพื่อน จึงได้ทำการลบแก้ไขใหม่ได้ดังนี้

```python
datan1 = datan[datan['Country']!='Mainland China']
datan1
```

```python
plt.bar(datan1.iloc[:,0], datan1.iloc[:,5], color = 'b')
plt.title(u'จำนวนผู้ติดเชื้อ covid-19 ของแต่ละประเทศ', fontname='Tahoma', fontsize='16', color = 'b')
plt.xlabel('Country', fontname='Tahoma', fontsize='16', color = 'b')
plt.ylabel('Confirmed', fontname='Tahoma', fontsize='16', color = 'b')
```

```python
plt.bar(datan1.iloc[:,0], datan1.iloc[:,7], color = 'g')
plt.title(u'จำนวนผู้ที่รักษา covid-19 หายแล้ว ของแต่ละประเทศ', fontname='Tahoma', fontsize='16', color = 'g')
plt.xlabel('Country', fontname='Tahoma', fontsize='16', color = 'g')
plt.ylabel('Recovered', fontname='Tahoma', fontsize='16', color = 'g')
```

```python
plt.bar(datan1.iloc[:,0], datan1.iloc[:,6], color = 'r')
plt.title(u'จำนวนผู้เสียชีวิตสาเหตุจาก covid-19 ของแต่ละประเทศ', fontname='Tahoma', fontsize='16', color = 'r')
plt.xlabel('Country', fontname='Tahoma', fontsize='16', color = 'r')
plt.ylabel('Deaths', fontname='Tahoma', fontsize='16', color = 'r')
```

# 5.เพื่อให้กราฟดูง่ายยิ่งขึ้น จึงทำการสุ่มตัวอย่างมาเปรียบเทียบดู จำนวน 8 ประเทศ 

```python
datan1_sample = datan1.sample(n=8)
datan1_sample
```

```python
plt.bar(datan1_sample.iloc[:,0], datan1_sample.iloc[:,5], color = 'b')
plt.title(u'จำนวนผู้ติดเชื้อ covid-19 ใน 8 ประเทศ', fontname='Tahoma', fontsize='16', color = 'b')
plt.xlabel('Country', fontname='Tahoma', fontsize='16', color = 'b')
plt.ylabel('Confirmed', fontname='Tahoma', fontsize='16', color = 'b')
```

```python
plt.bar(datan1_sample.iloc[:,0], datan1_sample.iloc[:,7], color = 'g') 
plt.title(u'จำนวนผู้ที่รักษา covid-19 หายแล้วของแต่ละประเทศ', fontname='Tahoma', fontsize='16', color = 'g')
plt.xlabel('Country', fontname='Tahoma', fontsize='16', color = 'g')
plt.ylabel('Recovered', fontname='Tahoma', fontsize='16', color = 'g')
```

```python
plt.bar(datan1_sample.iloc[:,0], datan1_sample.iloc[:,6], color = 'r')
plt.title(u'จำนวนผู้เสียชีวิตสาเหตุจาก covid-19 ใน 8 ประเทศ', fontname='Tahoma', fontsize='16', color = 'r')
plt.xlabel('Country', fontname='Tahoma', fontsize='16', color = 'r')
plt.ylabel('Deaths', fontname='Tahoma', fontsize='16', color = 'r')
```

# 6.เปรียบเทียบค่าของข้อมูลจำนวนของผู้ที่ติดเชื้อ ผู้ที่รักษาหาย และผู้ที่ตาย ใน 10 ประเทศที่สุ่มตัวอย่างมา

```python
plt.bar([1,2,3,4,5,6,7,8], datan1_sample.iloc[:,5],tick_label = datan1_sample.iloc[:,0])
plt.title('จำนวนผู้ติดเชื้อ covid-19 ใน 8 ประเทศ', fontname='Tahoma', fontsize='16')
plt.xlabel('country')
plt.ylabel('จำนวน(หน่วย:คน)', fontname='Tahoma', fontsize='16')
```

```python
plt.bar([1-(0.35/2),2-(0.35/2),3-(0.35/2),4-(0.35/2),5-(0.35/2),6-(0.35/2),7-(0.35/2),8-(0.35/2)],
        datan1_sample.iloc[:,5],width=0.35,tick_label=datan1_sample.iloc[:,0],color= 'b',label='Confirmed',align='edge')
plt.bar([1+(0.35/2),2+(0.35/2),3+(0.35/2),4+(0.35/2),5+(0.35/2),6+(0.35/2),7+(0.35/2),8+(0.35/2)],
        datan1_sample.iloc[:,6],width=0.35,tick_label=datan1_sample.iloc[:,0],color= 'r',label='Deaths',align='edge')
plt.bar([1+(2.5*0.35/2),2+(2.5*0.35/2),3+(2.5*0.35/2),4+(2.5*0.35/2),5+(2.5*0.35/2),6+(2.5*0.35/2),7+(2.5*0.35/2),8+(2.5*0.35/2)],
        datan1_sample.iloc[:,7],width=0.35,tick_label=datan1_sample.iloc[:,0],color= 'g',label='Recovered',align='edge')
plt.title('เปรียบเทียบจำนวนผู้ติดเชื้อ ผู้ที่รักษาหายแล้ว และผู้เสียชีวิต สาเหตุจาก  covid-19 ใน 8 ประเทศ', fontname='Tahoma', fontsize='16')
plt.xlabel('country')
plt.legend(loc=1)
plt.ylabel('จำนวน(หน่วย:คน)', fontname='Tahoma', fontsize='16')
```
