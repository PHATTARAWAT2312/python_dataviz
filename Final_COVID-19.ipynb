# 9. ต้องการดูระยะเวลากับยอดผู้ติดเชื้อ covid-19 ว่ามีแนวโน้มอย่างไร
#### - สร้างคอลัมน์ใหม่ที่มีชื่อว่า date_new โดยค่าในคอลัมน์ได้มาจากคอลัมน์ Date ที่ผ่านการลบปีและเวลาออก เพื่อให้ง่ายต่อการ plot กราฟ

```python
date_n = []
for i in range(data.shape[0]):
    x = data.iloc[i]['Date'].split(' ')
    x = x[0].replace('/2020','')
    date_n.append(x)
date_n
```

```python
data['Date_new'] = date_n
data
```

## 9.1 แนวโน้มจำนวนผู้ติดเชื้อ covid-19 ในประเทศไทย ตั้งแต่เดือน 22 ม.ค.- 17 ก.พ. 2563

```python
x = data[data['Country']=='Thailand']['Date_new']
y1 = data[data['Country']=='Thailand']['Confirmed']
y2 = data[data['Country']=='Thailand']['Deaths']
y3 = data[data['Country']=='Thailand']['Recovered']

plt.figure(figsize=[10,5],dpi=80)
plt.xticks(rotation = 90)
plt.plot(x,y1,'blue',label = "Confirmed",marker='*')
plt.plot(x,y2,'green',label = "Recovered",marker='*')
plt.plot(x,y3,'red',label = "Deaths",marker='*')
plt.legend()

plt.xlabel('วันที่', fontname='Tahoma',fontsize=16)
plt.ylabel('จำนวน(คน)', fontname='Tahoma',fontsize=16)
plt.title('รวมจำนวนผู้ติดเชื้อ covid-19 ในประเทศไทย ตั้งแต่เดือน 22 ม.ค.- 17 ก.พ. 2563',fontname='Tahoma',fontsize=16)
```

## 9.2 แนวโน้มจำนวนผู้ติดเชื้อ covid-19 ทั่วโลก ตั้งแต่เดือน 22 ม.ค.- 17 ก.พ. 2563

```python
x = data.groupby(['Date_new'], as_index=False).sum()['Date_new']
y1 = data.groupby(['Date_new'], as_index=False).sum()['Confirmed']
y2 = data.groupby(['Date_new'], as_index=False).sum()['Recovered']
y3 = data.groupby(['Date_new'], as_index=False).sum()['Deaths']

plt.figure(figsize=[10,5],dpi=80)
plt.xticks(rotation = 90)
plt.plot(x,y1,'blue',label = "Confirmed",marker='.')
plt.plot(x,y2,'green',label = "Recovered",marker='.')
plt.plot(x,y3,'red',label = "Deaths",marker='.')
plt.legend()

plt.xlabel('วันที่', fontname='Tahoma',fontsize=16)
plt.ylabel('จำนวน(คน)', fontname='Tahoma',fontsize=16)
plt.title('รวมจำนวนผู้ติดเชื้อ covid-19 ทั่วโลก ตั้งแต่เดือน 22 ม.ค.- 17 ก.พ. 2563',fontname='Tahoma',fontsize=16)
```
